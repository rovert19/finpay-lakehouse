#Creacion de Catalogo y Schemas

```python
%sql

CREATE CATALOG IF NOT EXISTS fintech_finpay;
```

```python
%sql

USE CATALOG fintech_finpay;

CREATE SCHEMA IF NOT EXISTS bronze;

CREATE SCHEMA IF NOT EXISTS silver;

CREATE SCHEMA IF NOT EXISTS gold;

CREATE SCHEMA IF NOT EXISTS observability;
```

#Creacion de Volumen para capa landing

```python
%sql

USE CATALOG fintech_finpay;

USE SCHEMA default;

CREATE VOLUME IF NOT EXISTS vol_landing;
```

```python
%python

VOLUME = '/Volumes/fintech_finpay/default/vol_landing'
USER_FOLDER = 'users'
MERCHANT_FOLDER = 'merchants'
TRANX_FOLDER = 'transactions'
METADATA_FOLDER = 'metadata'

dbutils.fs.mkdirs(f'{VOLUME}/{USER_FOLDER}')
dbutils.fs.mkdirs(f'{VOLUME}/{MERCHANT_FOLDER}')
dbutils.fs.mkdirs(f'{VOLUME}/{TRANX_FOLDER}')
dbutils.fs.mkdirs(f'{VOLUME}/{METADATA_FOLDER}')
```

# Permisos por rol

```python
%sql

GRANT USE CATALOG ON CATALOG fintech_finpay TO riesgo;

GRANT USE SCHEMA, SELECT ON SCHEMA fintech_finpay.silver TO riesgo;

GRANT USE SCHEMA, SELECT ON SCHEMA fintech_finpay.gold TO riesgo;
```

```python
%sql

GRANT USE CATALOG ON CATALOG fintech_finpay TO auditoria;

GRANT USE SCHEMA, SELECT ON SCHEMA fintech_finpay.gold TO auditoria;

GRANT USE SCHEMA, SELECT ON SCHEMA fintech_finpay.observability TO auditoria;
```

```python
%sql

GRANT USE CATALOG, USE SCHEMA, CREATE TABLE, MODIFY ON CATALOG fintech_finpay TO ingenieria;
```
